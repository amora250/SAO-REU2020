```python
xHI_min = 0.
xHI_max = 1.

xHI_grid = np.linspace(-1,2)
def log_prior(theta):
    """
    Returns log(prior) for a given parameter vector
    
    Parameters
    ----------
    theta: list, array-like
        List of parameters in the form [slope, intercept]
    
    Returns
    -------
    ln_prior: float
        Natural log of the prior probability function
    """
    xHI = theta
    
    if xHI_min < xHI < xHI_max:
        prob_xHI = 1. # Log prior (not worrying about normalization)
    else:
        prob_xHI = 0.

    ln_prior = np.log(prob_xHI)  # Log of the probability
    return ln_prior

plt.plot(xHI_grid, [np.exp(log_prior(xHI)) for xHI in xHI_grid])
```
